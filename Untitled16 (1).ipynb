```python
##########################################
#          IMPORT LIBRARIES
##########################################
import os, sys, shutil, timeit, multiprocessing as mp, warnings
import numpy as np, pandas as pd
import librosa, librosa.display, seaborn as sns, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from joblib import Parallel, delayed
import kagglehub

warnings.filterwarnings("ignore")
print("✅ Libraries imported successfully")

##########################################
#          DOWNLOAD DATASET
##########################################
print("\n📦 Downloading RAVDESS dataset...")
ravdess_dir = kagglehub.dataset_download("uwrfkaggler/ravdess-emotional-speech-audio")
print("Dataset downloaded to:", ravdess_dir)

# Collect audio paths and emotion labels
audio_files, emotions = [], []
emotion_map = {1:'neutral',2:'calm',3:'happy',4:'sad',5:'angry',6:'fear',7:'disgust',8:'surprise'}

for root, _, files in os.walk(ravdess_dir):
    for f in files:
        if f.endswith('.wav'):
            parts = f.split('-')
            if len(parts) > 2:
                emo = int(parts[2])
                emotions.append(emotion_map.get(emo))
                audio_files.append(os.path.join(root, f))

print(f"✅ Total audio files: {len(audio_files)}")
df = pd.DataFrame({"Path": audio_files, "Emotion": emotions})
print(df["Emotion"].value_counts())

##########################################
#          AUDIO AUGMENTATION
##########################################
def add_noise(data):
    noise_amp = 0.005 * np.random.uniform() * np.amax(data)
    return data + noise_amp * np.random.normal(size=data.shape[0])

def pitch_shift(data, sr, n_steps=2):
    return librosa.effects.pitch_shift(y=data, sr=sr, n_steps=n_steps)

def stretch_audio(data, rate=0.8):
    return librosa.effects.time_stretch(y=data, rate=rate)

def shift_audio(data):
    shift_range = int(np.random.uniform(-5, 5) * 1000)
    return np.roll(data, shift_range)

##########################################
#          FEATURE EXTRACTION
##########################################
def extract_features(data, sr):
    # Compute MFCCs
    mfcc = librosa.feature.mfcc(y=data, sr=sr, n_mfcc=40)
    return np.mean(mfcc.T, axis=0)

def augment_and_extract(path, duration=2.5, offset=0.6, sr=22050):
    try:
        data, sr = librosa.load(path, duration=duration, offset=offset, sr=sr)
        augmented = [
            data,
            add_noise(data),
            pitch_shift(data, sr),
            stretch_audio(data),
            shift_audio(data)
        ]
        return [extract_features(a, sr) for a in augmented]
    except Exception as e:
        print(f"⚠️ Error processing {path}: {e}")
        return [np.zeros(40)] * 5

##########################################
#          PARALLEL EXTRACTION
##########################################
start = timeit.default_timer()
print("\n🎧 Extracting MFCC features (with augmentation)...")

def process_file(path, emotion):
    feats = augment_and_extract(path)
    return feats, [emotion] * len(feats)

results = Parallel(n_jobs=mp.cpu_count())(delayed(process_file)(p, e) for p, e in zip(df['Path'], df['Emotion']))

X, Y = [], []
for feats, emos in results:
    X.extend(feats)
    Y.extend(emos)

stop = timeit.default_timer()
print(f"✅ Feature extraction completed in {round(stop-start,2)} seconds")
print("Total augmented samples:", len(X))

##########################################
#          DATA PREPARATION
##########################################
X = np.array(X)
Y = np.array(Y)

scaler = StandardScaler()
X = scaler.fit_transform(X)

ohe = OneHotEncoder(sparse_output=False)
Y_encoded = ohe.fit_transform(Y.reshape(-1,1))

x_train, x_test, y_train, y_test = train_test_split(
    X, Y_encoded, test_size=0.2, random_state=42, stratify=Y_encoded
)

x_train = np.expand_dims(x_train, axis=2)
x_test = np.expand_dims(x_test, axis=2)
print("x_train:", x_train.shape, "x_test:", x_test.shape)

##########################################
#          BUILD MODEL
##########################################
model = tf.keras.Sequential([
    Conv1D(256, 5, activation='relu', padding='same', input_shape=(x_train.shape[1], 1)),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.3),

    Conv1D(128, 5, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.3),

    Bidirectional(LSTM(128, return_sequences=True)),
    Bidirectional(LSTM(64)),

    Dense(128, activation='relu'),
    Dropout(0.4),
    Dense(Y_encoded.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

##########################################
#          TRAIN MODEL
##########################################
callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=5, min_lr=1e-6)
]

history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=20,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

##########################################
#          PLOTS
##########################################
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

##########################################
#          EVALUATION
##########################################
y_pred = np.argmax(model.predict(x_test), axis=1)
y_true = np.argmax(y_test, axis=1)
print("\n📊 Classification Report:\n", classification_report(y_true, y_pred, target_names=ohe.categories_[0]))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='viridis', xticklabels=ohe.categories_[0], yticklabels=ohe.categories_[0])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

##########################################
#          SAVE MODEL
##########################################
model.save("emotion_recognition_model_stable.h5")
print("✅ Model saved as 'emotion_recognition_model_stable.h5'")
```

```python


##########################################
#          TEST / PREDICTION PHASE
##########################################
from google.colab import files
import tensorflow as tf
import numpy as np
import librosa

# Load the trained model (match filename!)
model = tf.keras.models.load_model('emotion_recognition_model_stable.h5')

# Must match training labels
EMOTIONS = ['neutral', 'calm', 'happy', 'sad', 'angry', 'fear', 'disgust', 'surprise']

def extract_features(file_path, duration=2.5, offset=0.6, sr=22050):
    """
    Matches the same MFCC extraction used during training.
    Returns a (1, 40, 1) shaped array for prediction.
    """
    try:
        data, sr = librosa.load(file_path, duration=duration, offset=offset, sr=sr)
        mfcc = librosa.feature.mfcc(y=data, sr=sr, n_mfcc=40)
        mfcc_scaled = np.mean(mfcc.T, axis=0)  # same as training
        mfcc_scaled = mfcc_scaled.reshape(1, -1, 1)  # shape (1, 40, 1)
        return mfcc_scaled
    except Exception as e:
        print(f"Error extracting {file_path}: {e}")
        return np.zeros((1, 40, 1))

# Upload an audio file for testing
uploaded = files.upload()

# Predict emotion for each uploaded file
for file_name in uploaded.keys():
    try:
        features = extract_features(file_name)
        predictions = model.predict(features)
        predicted_emotion = EMOTIONS[np.argmax(predictions)]

        print(f"\n🎧 File: {file_name}")
        print(f"Predicted Emotion: {predicted_emotion}")
        print("Confidence scores:")
        for emotion, score in zip(EMOTIONS, predictions[0]):
            print(f"  {emotion}: {score:.4f}")

    except Exception as e:
        print(f"Error processing {file_name}: {e}")
```
